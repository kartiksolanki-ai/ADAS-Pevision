```python
import cv2
from ultralytics import YOLO
import numpy as np

# Function to issue a warning when a pedestrian is detected moving towards the car
#this function has been enabled only for the pedestrians coming towards the car and requires a specific dataset
def issue_warning(frame):
    height, width, _ = frame.shape
    warning_text = "Pedestrian detected moving towards the car! Issuing a warning."
    # Calculate position to display the warning message
    text_size, _ = cv2.getTextSize(warning_text, cv2.FONT_HERSHEY_PLAIN, 1, 1)
    text_x = int((width - text_size[0]) / 2)
    text_y = text_size[1] + 10  # Adjust this value as needed
    cv2.putText(frame, warning_text, (text_x, text_y), cv2.FONT_HERSHEY_PLAIN, 1, (0, 0, 255), 2)

cap = cv2.VideoCapture("data2.mp4")

model = YOLO("yolov8m.pt")

# Initialize a dictionary to store previous positions of each pedestrian
prev_positions = {}

# Flag to track if a warning has been issued
warning_issued = False

# Iterate for 10 epochs#it can be changed as well
for epoch in range(5):
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame, device="mps")

        # Iterate over each result in the list
        for result in results:
            bboxes = np.array(result.boxes.xyxy.cpu(), dtype="int")  # Accessing bounding boxes
            classes = np.array(result.boxes.cls.cpu(), dtype="int")  # Accessing class labels

            # Iterate over each detected pedestrian
            for bbox, cls in zip(bboxes, classes):
                (x, y, x2, y2) = bbox

                # Calculate centroid of the bounding box
                centroid_x = int((x + x2) / 2)
                centroid_y = int((y + y2) / 2)

                # Save current position
                person_id = str(cls)  # Assuming cls is unique for each person
                prev_positions.setdefault(person_id, []).append((centroid_x, centroid_y))

                # Predict the future direction of the pedestrian
                if person_id in prev_positions and len(prev_positions[person_id]) >= 2:
                    trajectory = prev_positions[person_id]
                    current_position = trajectory[-1]
                    previous_position = trajectory[-2]

                    # Determine the movement direction
                    horizontal_direction = "right" if current_position[0] > previous_position[0] else "left"
                    vertical_direction = "backward" if current_position[1] < previous_position[1] else "forward"

                    # Update bounding box color based on movement direction
                    if horizontal_direction == "right":
                        cv2.rectangle(frame, (x, y), (x2, y2), (0, 255, 255), 2)  # Yellow color for moving right
                    else:
                        cv2.rectangle(frame, (x, y), (x2, y2), (0, 0, 255), 2)  # Red color for moving left

                    if vertical_direction == "forward":
                        cv2.rectangle(frame, (x, y), (x2, y2), (0, 255, 255), 2)  # Yellow color for moving forward
                    else:
                        cv2.rectangle(frame, (x, y), (x2, y2), (255, 0, 0), 2)  # Blue color for moving backward

                    # Display the future direction with smaller font size
                    font_scale = 1  # Font scale for smaller text
                    movement_directions = [vertical_direction, horizontal_direction]
                    direction_text = ', '.join(movement_directions)
                    text_size, _ = cv2.getTextSize(direction_text, cv2.FONT_HERSHEY_PLAIN, font_scale, 2)
                    text_x = centroid_x - text_size[0] // 2
                    text_y = y - 10  # Adjust this value to control the distance between the text and the bounding box
                    cv2.putText(frame, direction_text, (text_x, text_y), cv2.FONT_HERSHEY_PLAIN, font_scale, (0, 255, 255), 2)

                    # Check if pedestrian is moving towards the car
                    if horizontal_direction == "right" and vertical_direction == "forward" and not warning_issued:
                        issue_warning(frame)
                        warning_issued = True

        cv2.imshow("ADASPe-Vision", frame)
        key = cv2.waitKey(1)
        if key == 27:
            break

    cap.set(cv2.CAP_PROP_POS_FRAMES, 0)  # Reset video to the beginning for the next epoch
    warning_issued = False  # Reset warning flag for the next epoch

cap.release()
cv2.destroyAllWindows()

```

```python
import cv2
from ultralytics import YOLO
import numpy as np
import math

# Function to smooth angles using a simple moving average
def smooth_angles(angles, window_size=5):
    smoothed_angles = []
    for i in range(len(angles)):
        start_idx = max(0, i - window_size // 2)
        end_idx = min(len(angles), i + window_size // 2 + 1)
        window = angles[start_idx:end_idx]
        smoothed_angle = np.mean(window)
        smoothed_angles.append(smoothed_angle)
    return smoothed_angles

# Load the video frames without running the video
cap = cv2.VideoCapture("data2.mp4")

# Initialize the YOLO model
model = YOLO("yolov8m.pt")

# Initialize variables to keep track of total angles and total pedestrians
total_angles = 0
total_pedestrians = 0

# Initialize dictionary to store previous positions of pedestrians
prev_positions = {}

# Initialize frame counter
frame_counter = 0

# Process each frame in the video
while True:
    ret, frame = cap.read()
    if not ret:
        break  # Break the loop when no more frames are available
    
    frame_counter += 1
    
    # Select every 10th frame
    if frame_counter % 10 != 0:
        continue
    
    # Run YOLO on the current frame to detect pedestrians
    results = model(frame, device="mps")

    # Iterate over each result in the list
    for result in results:
        bboxes = np.array(result.boxes.xyxy.cpu(), dtype="int")  # Accessing bounding boxes
        classes = np.array(result.boxes.cls.cpu(), dtype="int")   # Accessing class labels

        # Iterate over each detected pedestrian
        for bbox, cls in zip(bboxes, classes):
            (x, y, x2, y2) = bbox

            # Calculate centroid of the bounding box
            centroid_x = int((x + x2) / 2)
            centroid_y = int((y + y2) / 2)

            # Save current position
            person_id = str(cls)  # Assuming cls is unique for each person
            prev_positions.setdefault(person_id, []).append((centroid_x, centroid_y))
            total_pedestrians += 1

# Calculate average angular error if pedestrians were detected
if total_pedestrians > 0:
    angles = []
    for person_id, trajectory in prev_positions.items():
        if len(trajectory) >= 3:  # Ensure at least three positions are available
            for i in range(2, len(trajectory)):
                current_position = trajectory[i]
                previous_position = trajectory[i-1]

                # Calculate angle between predicted and actual positions
                angle = math.atan2(current_position[1] - previous_position[1], current_position[0] - previous_position[0])

                angles.append(angle)

    # Smooth angles
    smoothed_angles = smooth_angles(angles)

    # Calculate average angular error
    total_smoothed_angles = sum(smoothed_angles)
    average_error_degrees = (total_smoothed_angles / len(smoothed_angles)) * (180 / np.pi)
    print("Smoothed Average angular error (degrees):", average_error_degrees)

# Release the video capture object
cap.release()
```

```python
# Define the maximum allowable angular error in degrees
max_allowable_error_degrees = 40 # Adjust this according to your requirements

# Convert the average angular error to a percentage
error_percentage = (average_error_degrees / max_allowable_error_degrees) * 100

print("Error Percentage:", error_percentage)

```

```python
# Calculate accuracy by subtracting the error percentage from 100%
accuracy = 100 - error_percentage
print("Accuracy of the model:", accuracy, "%")
```
